```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import earthpy as et
from shapely.geometry import Point, Polygon
```

```python
def evalua_puntos(poligono,radiacion_global):
    puntos_dentro_poligono = []
    for i in range(len(radiacion_global)):
        punto = Point(radiacion_global.iloc[i,1],radiacion_global.iloc[i,0])
        if punto.within(poligono) == True:
            puntos_dentro_poligono.append(punto)
    
    return puntos_dentro_poligono
```

```python
def radiaciones_promedio(puntos,radiacion_global):
    radiaciones_puntos = []
    radiaciones_promedios =[]
    for i in range(len(puntos)):
        for punto in puntos[i]:
            LON = punto.x
            LAT = punto.y
            fila_seleccionada = radiacion_global.loc[(radiacion_global['LON'] == LON) & (radiacion_global['LAT'] == LAT)]
            radiacion_seleccionada = fila_seleccionada["ANN"]
            radiaciones_puntos.append(float(radiacion_seleccionada))
        radiaciones_promedios.append(sum(radiaciones_puntos) / len(radiaciones_puntos))
    
    return radiaciones_promedios
            
```
